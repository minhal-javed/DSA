# Constant Complexity (O(C))

```python
def constant(n):
    result1=n[0]*n[0]
    print(result1)
    
constant([2,3])

```

# Linear Complexity (O(n))

```python
def linear_algo(items):
    for item in items:
        print(item)

linear_algo([4, 5, 6, 8])
```

```python
x = [2, 4, 6, 8, 10, 12]

y = [2, 4, 6, 8, 10, 12]

plt.plot(x, y, 'b')
plt.xlabel('Inputs')
plt.ylabel('Steps')
plt.title('Linear Complexity')
plt.show()
```
